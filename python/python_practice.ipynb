```python
def scatterIrisData(data) : 
  """Retrieve Petal attributes from the dataset & scatter on a canvas."""
  plt.figure() # create canvas
  plt.scatter(data[:, 2], data[:, 3]) # scatter plot

  plt.title('Irish Dataset Visualization')
  plt.xlabel('Petal Length')
  plt.ylabel('Petal Width')


  plt.show() # display


# data_path = "/content/drive/My Drive/Colab Notebooks/Iris_data.npy"
data = np.load('Iris_data.npy') # load data from the file in session storage - temporary storage
print(f'Shape : {data.shape}')

scatterIrisData(data)


```
